```python
import torch
import torch.nn as nn

# 1. On définit la fonction de perte
criterion = nn.CrossEntropyLoss()

# 2. Imaginons 3 classes (Chat, Chien, Oiseau)
# Les scores bruts du modèle (logits) pour 1 image
input_scores = torch.randn(1, 3) 

# 3. La vraie étiquette (la classe "Chien" est l'indice 1)
target = torch.tensor([1])

# 4. Calcul de la perte
loss = criterion(input_scores, target)
print(f"Perte : {loss.item()}")
```
